## 1. المقدمة
المشروع عبارة عن نظام إدارة محتوى (CMS) بسيط وودود للمبتدئين، يهدف إلى إنشاء وإدارة **صفحات الموقع** و**المقالات المدونة**، شبيه بإصدار مصغّر جدًا من WordPress أو Drupal.  
يتكون النظام من موقع عام للزوار + لوحة تحكم داخلية لإدارة المحتوى، وقوائم التنقل، والإعدادات الأساسية للموقع.

## 2. أهداف المشروع
مساعدة المستخدمين غير التقنيين على:
- إنشاء ونشر صفحات جديدة بسهولة (نبذة عنا، تواصل معنا…).  
- كتابة وإدارة مقالات المدونة دون كتابة أي كود.  
- ترتيب الصفحات في قائمة تنقل عليا تظهر للزائر.  
- حفظ المحتوى كمسودّة والتحكم في النشر أو إلغاء النشر.  
- الحصول على موقع احترافي المظهر بأقل مجهود ممكن.

## 3. الجمهور المستهدف
- أصحاب الأعمال الصغيرة الذين يريدون إدارة محتوى موقعهم بأنفسهم  
- المدونون وصناع المحتوى  
- الجمعيات والمواقع الشخصية التي تحتاج إدارة صفحات بسيطة  
- كل من يبحث عن بديل سهل ومجاني للـ CMS المعقدة

## 4. الخصائص الرئيسية

### <strong>لوحة التحكم (Dashboard)</strong>
- نظرة عامة على المحتوى والنشاط الأخير  
- أزرار سريعة لإنشاء صفحة أو مقالة جديدة

### <strong>إدارة الصفحات</strong>
- إنشاء، تعديل، نشر، أو إلغاء نشر الصفحات  
- إمكانية إظهار الصفحة في القائمة العلوية أو إخفاؤها  
- حذف الصفحات (المدير فقط)

### <strong>إدارة مقالات المدونة</strong>
- كتابة وتعديل المقالات  
- حفظ كمسودّة أو نشر مباشرة  
- تصنيفات وعلامات (اختياري)  
- حذف المقالات (المدير فقط)

### <strong>إدارة القوائم (المدير فقط)</strong>
- إضافة الصفحات إلى القائمة العلوية  
- إعادة ترتيب العناصر بالسحب والإفلات  
- إخفاء صفحة من القائمة دون إلغاء نشرها

### <strong>إعدادات الموقع الأساسية (المدير فقط)</strong>
- تعديل عنوان الموقع والشعار الجانبي  
- رفع شعار (Logo)  
- اختيار الصفحة الرئيسية (صفحة ثابتة أو قائمة المقالات)

### <strong>أدوار المستخدمين</strong>
- **المدير (Admin)**: صلاحيات كاملة  
- **المحرر (Editor)**: يستطيع إدارة الصفحات والمقالات فقط  
- **الزائر العام**: يرى المحتوى المنشور فقط

### <strong>الموقع العام (Public Website)</strong>
- صفحة رئيسية نظيفة مع الشعار والقائمة العلوية  
- صفحات ثابتة منشورة  
- قسم المدونة مع المقالات المنشورة  
- صفحة منفصلة لكل صفحة أو مقالة

## 5. حزمة التقنيات (مرنة تمامًا)
- **الواجهة الأمامية**: HTML + CSS + JS، أو React، Vue، Svelte، أو حتى Flask + Jinja  
- **الخلفية**: أي لغة (Node.js/Express، Python Flask/Django، PHP Laravel، إلخ)  
- **قاعدة البيانات**: PostgreSQL (مفضل)، MySQL، SQLite، أو حتى JSON/Markdown files  

الهدف: أن يكون المشروع سهل الفهم والتوسعة حتى للمبتدئين.

## 6. هيكلية النظام

### <strong>الموقع العام</strong>
- يعرض فقط المحتوى المنشور  
- القائمة العلوية تعكس إعدادات لوحة التحكم

### <strong>لوحة التحكم الإدارية</strong>
- متاحة فقط للمستخدمين المسجلين  
- المدير لديه صلاحيات كاملة، المحرر صلاحيات محدودة  
- أقسام: الصفحات، المقالات، القوائم، الإعدادات

### <strong>سير عمل المحتوى</strong>
مسودّة → منشور → يظهر في الموقع العام

### <strong>الصلاحيات حسب الدور</strong>
- المدير: كل شيء  
- المحرر: المحتوى فقط  
- الزائر: قراءة فقط
