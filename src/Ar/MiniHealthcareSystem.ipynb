# نظام صحي مصغّر – دفتر الشروط

## <strong>1. المشكلة (السياق والتحدي)</strong>

يواجه المرضى في عصرنا الحالي صعوبة في إيجاد طريقة بسيطة وسريعة لحجز المواعيد الطبية.  
المنصات الحالية غالبًا ما تكون معقدة جدًا، تتطلب تسجيل حساب كامل، أو تقدم ميزات زائدة عن الحاجة لنظام صحي مصغّر.

كما يحتاج الأطباء المستقلون والعيادات الصغيرة إلى أداة خفيفة تتيح لهم استقبال طلبات الحجز بسهولة دون بنية تحتية ثقيلة.

### <strong>هدف المشروع</strong>
إنشاء **نظام حجز مواعيد صحية مصغّر (Mini Healthcare System)** يتيح للمستخدمين:
- تصفح الأطباء
- عرض تفاصيل كل طبيب
- اختيار موعد وهمي
- حجز الموعد عبر نموذج بسيط  
مع خلفية بسيطة تستقبل الطلبات فقط.

## <strong>2. الجمهور المستهدف</strong>

### <strong>المرضى</strong>
- يريدون حجز موعد طبي بسرعة وسهولة  
- يكفيهم معلومات أساسية عن الطبيب (التخصص، المواعيد المتاحة)  
- يفضلون واجهة سريعة بدون تسجيل إجباري

### <strong>الأطباء والعيادات</strong>
- يحتاجون إلى أداة خفيفة لاستقبال طلبات الحجز  
- لا يريدون نظام سجلات طبية إلكترونية معقد (EHR)

## <strong>3. أهداف المشروع</strong>

- توفير واجهة حجز مواعيد سهلة وسريعة  
- عرض قائمة الأطباء مع المعلومات الأساسية  
- صفحة تفاصيل الطبيب مع مواعيد وهمية للأيام القادمة  
- نموذج حجز بسيط (بدون حساب)  
- محاكاة المواعيد المتاحة (بيانات وهمية)  
- خلفية بسيطة تحتوي على `POST /appointments`  
- واجهة حديثة ومتجاوبة بالكامل  
- نموذج تفاعلي عالي الجودة على **Figma**

## <strong>4. الخصائص المدمجة</strong>

### <strong>الصفحة الرئيسية</strong>
- عرض ترحيبي جذاب  
- شعار مثل: <em>«صحتك بضغطة واحدة»</em>  
- زر مباشر لقائمة الأطباء

### <strong>قائمة الأطباء</strong>
عرض 4–6 أطباء، كل طبيب يحتوي على:  
- صورة شخصية  
- الاسم الكامل  
- التخصص  
- وصف مختصر  
- زر "عرض الملف الشخصي"

### <strong>صفحة ملف الطبيب</strong>
- صورة + اسم + تخصص  
- الخبرة، الموقع، نبذة قصيرة  
- جدول المواعيد المتاحة (الـ 7–10 أيام القادمة مع الساعات)  
- زر "احجز موعدًا"

### <strong>نموذج الحجز</strong>
نموذج بسيط (بدون تسجيل دخول):  
- الاسم الكامل  
- رقم الهاتف أو البريد الإلكتروني  
- التاريخ والساعة المختارة  
- سبب الزيارة (اختياري)  
- زر "تأكيد الحجز"

### <strong>صفحة التأكيد</strong>
- ملخص الحجز  
- صورة الطبيب + الاسم + التخصص  
- التاريخ والوقت ومعلومات المريض  
- رسالة نجاح + أزرار إضافة إلى التقويم (Google/Outlook)

## <strong>5. المخرجات (Deliverables)</strong>

### <strong>نموذج Figma تفاعلي</strong>
- تدفق كامل قابل للنقر  
- نسخة موبايل + دسكتوب  
- التنقل: الرئيسية → قائمة الأطباء → ملف الطبيب → نموذج الحجز → التأكيد

### <strong>واجهة Frontend جاهزة ووظيفية</strong>
- HTML/CSS أو React أو Vue أو Flutter Web  
- متجاوبة بالكامل  
- تصميم طبي احترافي ومريح

### <strong>Backend وظيفي بسيط</strong>
- API بسيطة (Node/Express، Flask، Django…)  
- استقبال وتخزين الحجوزات  
- حفظ البيانات في ملف JSON أو SQLite

### <strong>عرض تقديمي (5–7 دقائق)</strong>
- المشكلة والحل  
- عرض مباشر للنموذج + الموقع

## <strong>6. المتطلبات التقنية</strong>

### <strong>Frontend</strong>
- الصفحات: الرئيسية → قائمة الأطباء → ملف الطبيب → نموذج الحجز → التأكيد  
- تصميم متجاوب (Mobile-First)

### <strong>Backend</strong>
- REST API  
- بيانات وهمية للأطباء والمواعيد  
- تخزين الحجوزات

### <strong>قاعدة البيانات (اختر واحدة)</strong>
- ملف JSON  
- SQLite  
- ذاكرة مؤقتة (للعرض فقط)

### <strong>تصميم الواجهة</strong>
- ألوان طبية موثوقة (أزرق، أبيض، لمسات خضراء)  
- خطوط واضحة وقابلية قراءة عالية  
- سهولة الوصول والاستخدام

## <strong>7. الخلاصة</strong>

مشروع **النظام الصحي المصغّر** هو مشروع هاكاثون مثالي يجمع بين:

- واجهة جميلة واحترافية  
- تدفق مستخدم سلس وعملي  
- خلفية خفيفة ووظيفية  
- حالة استخدام حقيقية في المجال الصحي  
- إمكانية الإنجاز السريع والعرض المبهر
