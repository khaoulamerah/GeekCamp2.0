# نظام إدارة مدرسية مصغّر – دفتر الشروط

## <strong>1. المشكلة (السياق والتحدي)</strong>

تواجه المدارس والمراكز التكوينية الصغيرة والمؤسسات التعليمية الحاجة إلى أداة بسيطة لإدارة الطلاب والمواد الدراسية والعلامات.  
معظم الحلول الموجودة إما ثقيلة جدًا، أو باهظة الثمن، أو تحتوي على ميزات زائدة عن الحاجة بالنسبة للمشاريع الأكاثون أو المشاريع الطلابية.

### <strong>هدف المشروع</strong>
بناء **نظام إدارة مدرسية مصغّر (Mini School Management System)** يتيح للأساتذة والإداريين:
- عرض المواد الدراسية، إدارة الطلاب، إضافة طالب جديد، وعرض العلامات، بواجهة نظيفة وخلفية خفيفة تعتمد على ملفات JSON فقط.

## <strong>2. الجمهور المستهدف</strong>

### <strong>الأساتذة وإداريي المدارس</strong>
- يريدون رؤية سريعة للطلاب المسجلين والمواد  
- يحتاجون إلى إضافة طالب جديد بسهولة  
- يفضلون عرض العلامات بشكل واضح ومنظم

### <strong>الطلاب</strong>
- يرغبون في معرفة المواد المتاحة  
- يحتاجون إلى الاطلاع على علاماتهم بطريقة بسيطة  
- يفضلون واجهة واضحة وسهلة الاستخدام

## <strong>3. أهداف المشروع</strong>

- توفير أداة إدارة مدرسية خفيفة وحديثة  
- عرض قائمة المواد الدراسية  
- إدارة قائمة الطلاب (عرض + إضافة)  
- عرض العلامات في جدول واضح (بيانات وهمية)  
- نموذج بسيط لإضافة طالب جديد  
- خلفية بسيطة تعتمد ملفات JSON  
- واجهة تعليمية متجاوبة وجميلة  
- نموذج تفاعلي عالي الجودة على **Figma**

## <strong>4. الخصائص المدمجة</strong>

### <strong>الصفحة الرئيسية</strong>
- رسالة ترحيب + شعار مثل: <em>«أدر مدرستك ببساطة»</em>  
- بطاقات وصول سريع إلى: المواد، الطلاب، العلامات، إضافة طالب

### <strong>قائمة المواد الدراسية</strong>
- عنوان المادة  
- رمز المادة (مثل: MATH101)  
- وصف مختصر  
- اسم الأستاذ (اختياري)  
- عدد الطلاب المسجلين

### <strong>قائمة الطلاب</strong>
- الاسم الكامل  
- رقم التسجيل  
- البريد الإلكتروني  
- المواد المسجل فيها  
- زر «عرض العلامات» (اختياري)

### <strong>نموذج إضافة طالب</strong>
- الاسم الكامل  
- البريد الإلكتروني  
- رقم التسجيل (تلقائي أو يدوي)  
- اختيار المادة/المواد  
- زر «إضافة الطالب»

### <strong>صفحة العلامات</strong>
جدول يعرض العلامات حسب الطالب والمادة:
- اسم الطالب  
- المادة  
- العلامة (مثل: A، B+، 85/100…)  
- الحالة (ناجح / راسب)

## <strong>5. المخرجات (Deliverables)</strong>

### <strong>نموذج Figma تفاعلي</strong>
- تدفق كامل قابل للنقر  
- نسخة موبايل + دسكتوب  
- التنقل: الرئيسية → المواد → الطلاب → إضافة طالب → العلامات

### <strong>واجهة Frontend جاهزة ووظيفية</strong>
- HTML/CSS أو React أو Vue أو Flutter Web  
- متجاوبة بالكامل  
- تصميم تعليمي أنيق (أزرق فاتح، أخضر، رمادي فاتح)

### <strong>Backend وظيفي بسيط</strong>
- API بسيطة (Node/Express، Flask، Django…)  
- قراءة وكتابة في ملفات JSON  
- عمليات CRUD بسيطة للطلاب

### <strong>عرض تقديمي (5–7 دقائق)</strong>
- المشكلة والحل  
- عرض مباشر لنموذج Figma + التطبيق العامل

## <strong>6. المتطلبات التقنية</strong>

### <strong>Frontend</strong>
- الصفحات: الرئيسية، المواد، قائمة الطلاب، إضافة طالب، العلامات  
- تصميم متجاوب (Mobile-First)  
- أسلوب تعليمي احترافي

### <strong>Backend</strong>
- REST API تعمل مع ملفات JSON  
- ملفات البيانات:  
  → `modules.json`  
  → `students.json`  
  → `grades.json` (اختياري)

### <strong>قاعدة البيانات</strong>
- ملفات JSON فقط (لا حاجة لقاعدة بيانات حقيقية)  
- أو SQLite إذا أردت

### <strong>تصميم الواجهة</strong>
- لوحة ألوان تعليمية (أزرق، أبيض، أخضر فاتح)  
- جداول وبطاقات واضحة  
- أيقونات تعليمية (قبعة تخرج، كتاب، قلم…)

## <strong>7. الخلاصة</strong>

مشروع **النظام الإداري المدرسي المصغّر** هو فكرة مثالية للهاكاثون أو المشاريع الجامعية، لأنه يجمع بطريقة متوازنة بين:

- واجهة جميلة وبسيطة  
- هيكلة Frontend عملية  
- خلفية خفيفة لكن وظيفية  
- حالة استخدام تعليمية حقيقية  
- سرعة التطوير والعرض المبهر
