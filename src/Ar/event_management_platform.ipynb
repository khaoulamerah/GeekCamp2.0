# مواصفات منصة إدارة الفعاليات الجامعية

## <strong>المقدمة - نظرة عامة على المشروع</strong>

يتضمن المشروع تطوير تطبيق ويب أو موبايل يركّز جميع الفعاليات التي تنظمها النوادي الجامعية في مكان واحد.  
يتيح التطبيق للنوادي الجامعية نشر فعالياتها، إدارة التسجيلات، معاينة المشاركين، كما يتيح للطلاب التسجيل بسهولة في هذه الفعاليات.

## <strong>الجمهور المستهدف</strong>

تستهدف المنصة:

- الطلاب الجامعيين الراغبين في تصفح الفعاليات، الاطلاع على تفاصيلها، والتسجيل فيها.  
- مسيّري النوادي الجامعية الذين يحتاجون إلى إنشاء الفعاليات وإدارتها ومتابعة قوائم المشاركين.

## <strong>الميزات الرئيسية</strong>

### <strong>الميزات الأساسية (MVP - المستوى السهل)</strong>

#### <strong>بالنسبة للطلاب (المستخدمين العاديين)</strong>
- تصفح قائمة الفعاليات المتاحة.  
- عرض تفاصيل الفعالية (العنوان، التاريخ والوقت، المكان، الوصف، عدد الأماكن المتبقية).  
- التسجيل في الفعالية.  
- إلغاء التسجيل.  
- عرض قائمة التسجيلات الشخصية في صفحة الملف الشخصي.

#### <strong>بالنسبة للمسؤولين (النوادي الجامعية)</strong>
- إضافة فعالية جديدة.  
- تعديل أو حذف فعالية.  
- تحديد عدد الأماكن المتاحة لكل فعالية.  
- معاينة قائمة المشاركين.  
- قبول مشارك (تغيير حالته إلى "مؤكد/تم التحقق منه").  
- حذف مشارك (إلغاء تسجيله).

#### <strong>ميزات عامة وأساسية</strong>
- مصادقة بسيطة باستخدام البريد الإلكتروني وكلمة المرور.  
- قاعدة بيانات لحفظ المستخدمين والفعاليات والتسجيلات.  
- تصفية الفعاليات حسب التاريخ، التصنيف، والنادي المنظم.

### <strong>ميزات إضافية</strong>

لم يتم تحديد ميزات إضافية خارج الوظائف الأساسية.  
تختلف مستويات الصعوبة بشكل أساسي في تكنولوجيا التطوير (مبتدئ مقابل متقدم).

## <strong>المخرجات</strong>

- مخططات UML (مخطط حالات الاستخدام، مخطط الكلاسات، مخطط التسلسل) – اختياري.  
- نماذج واجهة المستخدم (Mockups) باستخدام Figma أو رسومات يدوية.  
- الكود المصدري كاملاً.  
- عرض تقديمي نهائي للمشروع.

## <strong>المواصفات التقنية</strong>

- <strong>المتطلبات غير الوظيفية</strong>: واجهة بسيطة، تحميل سريع، أمان أساسي (تشفير كلمات المرور)، كود منظم وموثق.

### <strong>للمبتدئين (المستوى السهل)</strong>
- الواجهة الأمامية: HTML / CSS / JavaScript  
- الخلفية: PHP أو Node.js بسيط  
- قاعدة البيانات: MySQL

### <strong>للمتقدمين (المستوى الصعب)</strong>
- الواجهة الأمامية: React أو Flutter  
- الخلفية: Node.js، Laravel، أو Django  
- قاعدة البيانات: MySQL أو MongoDB
