## Introduction - Project Overview

The project involves developing a web or mobile application that centralizes all events organized by university clubs. It enables university clubs to publish their events, manage registrations, view participants, and allows students to register easily for these events.

## Target Audience

This platform is targeted at university students who wish to browse, view details, and register for events, as well as university club administrators who need to create, manage, and oversee events and participant lists.

## Key Features

### Core Features (MVP - Easy Level)
- **For Students (Users)**:
  - Browse the list of available events.
  - View event details including title, date and time, location, description, and number of remaining places.
  - Register for an event.
  - Cancel a registration.
  - View a list of their own registrations on a profile page.
- **For Administrators (University Clubs)**:
  - Add a new event.
  - Edit or delete an event.
  - Define the number of places available for an event.
  - View the list of participants.
  - Accept a participant (set status to "validated").
  - Delete a participant (cancel their registration).
- Simple authentication using email and password.
- Database storage for users, events, and registrations.
- Event filtering by date, category, and club.

### Additional Features
No additional features are specified beyond the core functionalities. Complexity levels are primarily differentiated in the technology stack (beginner vs. advanced implementations).

## Technical Specifications
- **Non-Functional Requirements**: Simple interface, fast loading, basic security (password hashing), and structured, documented code.
- **For Beginners (Easy Level)**:
  - Frontend: HTML/CSS/JS.
  - Backend: PHP or simple Node.js.
  - Database: MySQL.
- **For Advanced (Hard Level)**:
  - Frontend: React or Flutter.
  - Backend: Node.js, Laravel, or Django.
  - Database: MySQL or MongoDB.
