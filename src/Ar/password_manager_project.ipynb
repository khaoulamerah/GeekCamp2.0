# مواصفات مدير كلمات المرور الخفيف (Password Manager Lite)

## <strong>المقدمة - نظرة عامة على المشروع</strong>

يتكون المشروع من تطوير مدير كلمات مرور بسيط وآمن وعملي بالكامل، يتيح للمستخدمين إنشاء كلمات المرور، تخزينها، تنظيمها واسترجاعها سواء محليًا (offline) أو عبر الإنترنت (حسب مستوى الصعوبة).  
يركز المشروع على أفضل ممارسات الأمان (التجزئة، التشفير)، سهولة الاستخدام، والقابلية للنقل، مع الحفاظ على ملاءمته كمشروع تعليمي.  
الهدف هو توفير أداة عملية تعلّم عادات كلمات المرور الآمنة دون تعقيد الحلول التجارية مثل LastPass أو Bitwarden.

## <strong>الجمهور المستهدف</strong>

- طلاب ومطورون يتعلمون التشفير والتخزين الآمن.  
- أفراد يبحثون عن بديل خفيف وشفاف لمديري كلمات المرور التجارية.  
- مستخدمون يريدون التحكم الكامل في بياناتهم مع إمكانية العمل دون اتصال أولاً.  
- مدرسون ونوادي برمجة يستخدمونه كنموذج عملي لمشروع أمني.

## <strong>الميزات الرئيسية</strong>

### <strong>الميزات الأساسية (MVP - المستوى السهل)</strong>
منتج أدنى قابل للتطبيق – يعمل بالكامل دون اتصال وبدون خلفية:

- حماية بكلمة مرور رئيسية (Master Password) باستخدام PBKDF2 أو Argon2 (لا تُخزن أبدًا).  
- إضافة، تعديل وحذف المدخلات (اسم الموقع/التطبيق، اسم المستخدم/الإيميل، كلمة المرور، ملاحظات اختيارية).  
- مولد كلمات مرور آمن (طول قابل للتخصيص، أنواع الأحرف).  
- عرض قائمة بسيط مع بحث/تصفية حسب اسم الموقع.  
- نسخ كلمة المرور إلى الحافظة مع مسح تلقائي بعد 30 ثانية.  
- تشفير جميع البيانات من طرف العميل وحفظها في LocalStorage أو كملف JSON قابل للتنزيل.  
- واجهة نظيفة، متجاوبة، وبسيطة.

### <strong>المستوى المتوسط</strong>

- كل ميزات المستوى السهل.  
- قوالب خزائن متعددة (شخصي، عمل، عائلة...).  
- تصدير الخزينة المشفرة كملف (.json) + استيرادها.  
- توليد رمز QR لمشاركة مدخلة واحدة بأمان.  
- مقياس قوة كلمة المرور مع رد فعل بصري.  
- نظام تذكير بنسخ احتياطي محلي.  
- وضع الدارك مود وتحسين واجهة المستخدم/تجربة المستخدم.  
- دعم اختياري للمصادقة الثنائية (TOTP) للمدخلات.

### <strong>المستوى الصعب (متقدم)</strong>

- كل الميزات من المستويين السابقين.  
- نظام مصادقة كامل للمستخدمين (تسجيل/تسجيل دخول) مع إدارة جلسات آمنة.  
- خلفية + قاعدة بيانات مشفرة حقيقية (كلمات المرور لا تُخزن نصًا عاديًا أبدًا).  
- لوحة إدارة لإدارة المستخدمين (وضع عائلي أو متعدد المستخدمين).  
- مزامنة عبر الأجهزة عبر تخزين سحابي مشفر (Firebase أو مخصص).  
- مسح رمز QR (كاميرا ويب/موبايل) لاستيراد المدخلات.  
- قوالب خزينة قابلة للتخصيص (ألوان، أيقونات، تصنيفات).  
- تصدير تقرير ملخص الخزينة بصيغة PDF عالية الجودة (كلمات المرور مخفية).  
- دمج فحص التسريبات (HaveIBeenPwned API باستخدام k-anonymity).  
- واجهة احترافية مع حركات وإحساس تطبيق موبايل.  
- تقرير صحة كلمات المرور مدعوم بالذكاء الاصطناعي (اختياري) لاكتشاف الكلمات الضعيفة/المكررة.

## <strong>المخرجات</strong>

- الكود المصدري مستضاف على GitHub.  
- فيديو توضيحي مدته دقيقة واحدة يظهر إنشاء الخزينة، توليد كلمة مرور، والتصدير الآمن.  
- عرض تقديمي PDF يشرح خيارات الأمان (التجزئة، التشفير، أمان الحافظة).  
- ملف خزينة مشفر كنموذج + عرض عملية الاستعادة.

## <strong>المواصفات التقنية</strong>

- <strong>الواجهة الأمامية</strong>: HTML/CSS/JS، React، Vue.js، Flutter Web، أو Svelte.  
- <strong>الخلفية</strong> (متوسط/صعب): Node.js، Python Flask/Django، PHP Laravel، أو Firebase.  
- <strong>قاعدة البيانات</strong> (صعب): PostgreSQL، MySQL، MongoDB، أو Firebase Firestore (مع تشفير على مستوى الحقول).  

### <strong>التشفير</strong>
- كلمة المرور الرئيسية → اشتقاق مفتاح التشفير باستخدام Argon2id أو PBKDF2-SHA256.  
- تشفير الخزينة بـ AES-256-GCM.  
- لا تُخزن كلمة المرور الرئيسية أو مفتاح فك التشفير أبدًا.

### <strong>أدوات التصدير</strong>
jsPDF، html2canvas، FileSaver.js.

### <strong>أخرى</strong>
- يعمل بالكامل دون اتصال في المستوى السهل.  
- تصميم متجاوب (موبايل + حاسوب).  
- لا يتم الاحتفاظ بكلمات المرور كنص عادي في الذاكرة أكثر من اللازم.  
- رؤوس HTTP آمنة و Content Security Policy في الإنتاج.

<strong>المشروع تعليمي بامتياز، جذاب بصريًا في العروض (خاصة مولد كلمات المرور وتحريك فتح الخزينة)، ويعلّم مفاهيم الأمن السيبراني الحقيقية بمسؤولية.</strong>
