# مواصفات مفسر رسائل الأخطاء
## المقدمة - نظرة عامة على المشروع
المشروع عبارة عن تطبيق ويب بسيط مصمم لشرح رسائل الأخطاء الشائعة في لغة البرمجة C للمبتدئين، مع اقتراح حلول لها. يتيح التطبيق للمستخدم إدخال رسالة الخطأ ليحصل على شرح واضح يشمل الأسباب والحلول وأمثلة على الأكواد. يجب بناء التطبيق باستخدام HTML وCSS وJavaScript فقط (بدون خلفية)، وأن يعمل بالكامل دون اتصال بالإنترنت. الوثيقة الثانية تحتوي على أفكار تحسين بسيطة لرفع مستوى احترافية التطبيق دون تعقيد زائد.

## الجمهور المستهدف
يستهدف التطبيق المبرمجين المبتدئين في لغة C، مثل طلاب نوادي البرمجة أو المعسكرات البرمجية (مثل Geek Camp أو نادي الإعلام الآلي)، الذين يواجهون أخطاء شائعة أثناء الترجمة أو الربط أو التنفيذ ويحتاجون إلى شروحات مبسطة وإرشادات واضحة لحلها.

## الميزات الرئيسية
### الميزات الأساسية (MVP - المستوى السهل)
تمثل هذه المنتج الأدنى القابل للتطبيق بوظائف أساسية:
- حقل نصي (Textarea) للصق أو كتابة رسالة الخطأ، مع زر "شرح الخطأ" واختيارياً قائمة منسدلة لاختيار أخطاء شائعة.
- بحث غير حساس لحالة الأحرف يعتمد على الكلمات المفتاحية.
- عرض النتيجة على شكل بطاقة تحتوي على: عنوان الخطأ، شرح بسيط (جملتان إلى ثلاث)، الأسباب الشائعة (2-4 نقاط)، كيفية الإصلاح (2-5 خطوات)، وأمثلة قصيرة لكود سيء وصحيح.
- معالجة الأخطاء غير المعروفة بعرض رسالة "لم يتم العثور على خطأ مطابق" مع اقتراح كلمات مفتاحية مثل expected، undeclared، segmentation، undefined reference، format.
- قاعدة بيانات الأخطاء مخزنة داخل app.js كمصفوفة أو JSON، تحتوي على 15 خطأ على الأقل (منها الأخطاء المطلوبة مثل: expected ';'، undeclared variable، implicit declaration، format mismatch، expected ')'، expected '}'، segmentation fault، array bounds warning، undefined reference، non-void function control reaches end، scanf without '&').
- واجهة مستخدم بسيطة، واضحة، متجاوبة (للهواتف والحواسيب)، تعطي النتائج فوراً، وكود نظيف.

### ميزات إضافية
تعتمد على الميزات الأساسية وتُصنف حسب مستوى الصعوبة (بناءً على المكافآت الاختيارية وأفكار التحسين):
#### المستوى المتوسط (ميزات المكافأة)
- زر نسخ الحل لتسهيل نسخ الإصلاحات.
- اقتراح أقرب النتائج في حالة التطابق الجزئي.
- خاصية المفضلة باستخدام LocalStorage لحفظ الأخطاء الشائعة.
- خيارات تصفية حسب نوع الخطأ (ترجمة أو تشغيل).
- وضع الدارك مود (الوضع الليلي).

#### المستوى الصعب (تحسينات متقدمة)
- عرض أفضل 3 تطابقات مع نسب بسيطة (مثل 80% / 60% / 40%) للأخطاء غير المتطابقة تماماً.
- شارات توضح نوع الخطأ (Compile Error، Linker Error، Runtime Error).
- زر "إصلاح سريع" يعطي حلّاً مختصراً (خطوتين كحد أقصى) مع قسم "مزيد من التفاصيل" قابل للفتح.
- إبراز الكلمات المفتاحية في نص الخطأ المدخل (مثل expected، undeclared، segmentation).
- اقتراحات ذكية للأخطاء غير المعروفة (مثل اقتراح expected ';' أو undeclared).
- وضعان: مبتدئ (شروحات مبسطة) ومرشد (تفاصيل تقنية أعمق).
- وضع تعليمي صغير بعنوان "خطأ اليوم" مع شرح وحل قصير.

## المخرجات
- الكود المصدري (مستضاف على GitHub أو ما شابه).
- فيديو توضيحي مدته دقيقة واحدة يعرض التطبيق.
- عرض تقديمي بصيغة PDF للمشروع.
- مثال لشرح رسالة خطأ (لقطة شاشة أو نتيجة مُصدَّرة).

## المواصفات التقنية
- **الواجهة الأمامية**: HTML، CSS، JavaScript فقط.
- **الخلفية**: لا يوجد (تطبيق غير متصل بالإنترنت).
- **تخزين البيانات**: قاعدة بيانات الأخطاء كـ JSON أو مصفوفة في app.js؛ LocalStorage للمفضلة أو الميزات الدائمة.
- **أخرى**: يجب أن يكون متجاوباً، سريعاً، يعمل بدون إنترنت، بدون مكتبات خارجية أو ميزات تتطلب الاتصال.
